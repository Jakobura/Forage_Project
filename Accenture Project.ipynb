```python
import pandas as pd

# Load datasets
df1 = pd.read_csv("Cleaned/Content.csv")
df2 = pd.read_csv("Cleaned/Reactions.csv")
df3 = pd.read_csv("Cleaned/Reaction Types.csv")

# Ensure column names are stripped of extra spaces
df1.columns = df1.columns.str.strip()
df2.columns = df2.columns.str.strip()
df3.columns = df3.columns.str.strip()

# Merging df1 and df2 using an inner join on 'Content ID'
merged_df = pd.merge(df1, df2, on='Content ID', how='inner')

# Merging the result with df3 using an inner join on 'Reaction Type'
merged_df = pd.merge(merged_df, df3, on='Type', how='inner')

# Count the number of reactions per category
category_counts = merged_df.groupby('Category')['Type'].count().reset_index()

# Rename column for clarity
category_counts.rename(columns={'Type': 'total_reactions'}, inplace=True)

# Display result
print(category_counts)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load datasets
df1 = pd.read_csv("Cleaned/Content.csv")
df2 = pd.read_csv("Cleaned/Reactions.csv")
df3 = pd.read_csv("Cleaned/Reaction Types.csv")

# Ensure column names are stripped of extra spaces
df1.columns = df1.columns.str.strip()
df2.columns = df2.columns.str.strip()
df3.columns = df3.columns.str.strip()

# Merging df1 and df2 using an inner join on 'Content ID'
merged_df = pd.merge(df1, df2, on='Content ID', how='inner')

# Merging the result with df3 using an inner join on 'Reaction Type'
merged_df = pd.merge(merged_df, df3, on='Type', how='inner')

# Count the number of reactions per category
category_counts = merged_df.groupby('Category')['Type'].count().reset_index()

# Rename column for clarity
category_counts.rename(columns={'Type': 'total_reactions'}, inplace=True)

# Display result
print(category_counts)

# Plot a bar chart for total reactions per category
plt.figure(figsize=(10, 6))
plt.bar(category_counts['Category'], category_counts['total_reactions'], color='skyblue')

# Add titles and labels
plt.title('Total Reactions per Category')
plt.xlabel('Category')
plt.ylabel('Total Reactions')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
df1 = pd.read_csv("Cleaned/Content.csv")
df2 = pd.read_csv("Cleaned/Reactions.csv")
df3 = pd.read_csv("Cleaned/Reaction Types.csv")

# Ensure column names are stripped of extra spaces
df1.columns = df1.columns.str.strip()
df2.columns = df2.columns.str.strip()
df3.columns = df3.columns.str.strip()

# Merging df1 and df2 using an inner join on 'Content ID'
merged_df = pd.merge(df1, df2, on='Content ID', how='inner')

# Merging the result with df3 using an inner join on 'Reaction Type'
merged_df = pd.merge(merged_df, df3, on='Type', how='inner')

# Convert 'Datetime' to datetime format if not already
merged_df['Datetime'] = pd.to_datetime(merged_df['Datetime'])

# Extract month from 'Datetime'
merged_df['Month'] = merged_df['Datetime'].dt.month

# Count the number of reactions per month
month_counts = merged_df.groupby('Month')['Type'].count().reset_index()

# Rename the column for clarity
month_counts.rename(columns={'Type': 'total_posts'}, inplace=True)

# Create a pivot table for heatmap
heatmap_data = month_counts.pivot_table(index='Month', values='total_posts', aggfunc='sum')

# Plotting the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(heatmap_data, annot=True, cmap='Blues', cbar_kws={'label': 'Total Posts'}, linewidths=1, linecolor='black')

# Add titles and labels
plt.title('Reactions per Month Heatmap')
plt.xlabel('Month')
plt.ylabel('Total Posts')

# Show plot
plt.show()

```
